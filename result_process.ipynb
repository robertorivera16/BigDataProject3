```python
import csv
import re
import string
import json

tw_dict = []
with open("fetched_tweets.txt") as f:
    for i in range(2000):
        line = f.readline()
        tweet = json.loads(line)
        full_text = tweet['text']

        full_text = re.sub(r'[.,"!]+', '', full_text)  # removes the characters specified
        full_text = re.sub(r'^RT[\s]+', '', full_text)  # removes RT
        full_text = re.sub(r'https?:\/\/.[\r\n]', '', full_text)  # remove link
        full_text = re.sub(r'[:]+', '', full_text)
        new_line = ''
        for i in full_text.split():  # remove @ and #words, punctuataion
            if not i.startswith('@') and not i.startswith('#') and i \
                    not in string.punctuation:
                new_line += i + ' '
        full_text = new_line.lower()
        tw_dict.append(full_text)

result_dict = []

with open('test_results.txt', 'r') as results:
    for i in range(2000):
        entry = []
        entry.append(tw_dict[i])
        line = results.readline()
        entry.append(line[0])
        result_dict.append(entry)

flucounter = [0, 0, 0]
zikacounter = [0, 0, 0]
headachecounter = [0, 0, 0]
measlescounter = [0, 0, 0]
ebolacounter = [0, 0, 0]
diarrheacounter = [0, 0, 0]
        
for elem in result_dict:
    if (elem[0].find("flu") > -1):
        if (elem[1] == '0'):
            flucounter[0] += 1
        elif (elem[1] == '1'):
            flucounter[1] += 1
        elif (elem[1] == '2'):
            flucounter[2] += 1
    elif (elem[0].find("zika") > -1):
        if (elem[1] == '0'):
            zikacounter[0] += 1
        elif (elem[1] == '1'):
            zikacounter[1] += 1
        elif (elem[1] == '2'):
            zikacounter[2] += 1
    elif (elem[0].find("headache") > -1):
        if (elem[1] == '0'):
            headachecounter[0] += 1
        elif (elem[1] == '1'):
            headachecounter[1] += 1
        elif (elem[1] == '2'):
            headachecounter[2] += 1
    elif (elem[0].find("measles") > -1):
        if (elem[1] == '0'):
            measlescounter[0] += 1
        elif (elem[1] == '1'):
            measlescounter[1] += 1
        elif (elem[1] == '2'):
            measlescounter[2] += 1
    elif (elem[0].find("ebola") > -1):
        if (elem[1] == '0'):
            ebolacounter[0] += 1
        elif (elem[1] == '1'):
            ebolacounter[1] += 1
        elif (elem[1] == '2'):
            ebolacounter[2] += 1
    elif (elem[0].find("diarrhea") > -1):
        if (elem[1] == '0'):
            diarrheacounter[0] += 1
        elif (elem[1] == '1'):
            diarrheacounter[1] += 1
        elif (elem[1] == '2'):
            diarrheacounter[2] += 1
print()
print("Flu values : ", flucounter)
print("Zika values : ", zikacounter)
print("Headache values : ", headachecounter)
print("Measles values : ", measlescounter)
print("Ebola values : ", ebolacounter)
print("Diarrhea values : ", diarrheacounter)
print()
        
```
